```python
import matplotlib.pyplot as plt

# Data for charts
# Chart 1: Efficiency Comparisons
efficiency_data = {
    'Cell Type': ['Perovskite', 'Silicon'],
    'Efficiency (%)': [20, 17.5]
}

# Chart 2: Cost Comparison and Reduction Potential
cost_reduction_data = {
    'Aspect': ['Production Cost Reduction', 'Installation Cost Reduction'],
    'Reduction (%)': [25, 15]
}

# Chart 3: Environmental and Resource Considerations
env_impact_data = {
    'Aspect': ['Potential for Lead-Free', 'Resource Intensity Reduction', 'Recycling Potential'],
    'Score': [80, 60, 75]  # Arbitrary scores for illustration
}

# Chart 4: Performance in Diverse Lighting Conditions
performance_data = {
    'Condition': ['Low Light', 'High Temperature'],
    'Perovskite Performance': [85, 80],
    'Silicon Performance': [70, 60]
}

# Chart 5: Commercialization Challenges and Advancements
commercial_potential_data = {
    'Challenges': ['Stability', 'Pilot Projects', 'Policy Support'],
    'Progress (%)': [50, 70, 60]  # Arbitrary scores for illustration
}

# Function to plot green-shaded charts
def plot_green_charts():
    fig, axes = plt.subplots(3, 2, figsize=(12, 15))
    fig.suptitle("Results: Perovskite Solar Cells in India", fontsize=16, fontweight='bold')

    # Efficiency Comparisons
    axes[0, 0].bar(efficiency_data['Cell Type'], efficiency_data['Efficiency (%)'], color='darkgreen')
    axes[0, 0].set_title("Efficiency Comparisons")
    axes[0, 0].set_ylabel("Efficiency (%)")
    axes[0, 0].set_ylim(0, 25)

    # Cost Comparison and Reduction Potential
    axes[0, 1].bar(cost_reduction_data['Aspect'], cost_reduction_data['Reduction (%)'], color='green')
    axes[0, 1].set_title("Cost Comparison and Reduction Potential")
    axes[0, 1].set_ylabel("Reduction (%)")
    axes[0, 1].set_ylim(0, 30)

    # Environmental and Resource Considerations
    axes[1, 0].bar(env_impact_data['Aspect'], env_impact_data['Score'], color='seagreen')
    axes[1, 0].set_title("Environmental and Resource Considerations")
    axes[1, 0].set_ylabel("Impact Score (out of 100)")
    axes[1, 0].set_ylim(0, 100)

    # Performance in Diverse Lighting Conditions
    axes[1, 1].plot(performance_data['Condition'], performance_data['Perovskite Performance'], color='darkgreen', marker='o', label='Perovskite')
    axes[1, 1].plot(performance_data['Condition'], performance_data['Silicon Performance'], color='lightgreen', marker='s', linestyle='--', label='Silicon')
    axes[1, 1].set_title("Performance in Diverse Lighting Conditions")
    axes[1, 1].set_ylabel("Performance Score (out of 100)")
    axes[1, 1].legend()

    # Commercialization Challenges and Advancements
    axes[2, 0].bar(commercial_potential_data['Challenges'], commercial_potential_data['Progress (%)'], color='limegreen')
    axes[2, 0].set_title("Commercialization Challenges and Advancements")
    axes[2, 0].set_ylabel("Progress (%)")
    axes[2, 0].set_ylim(0, 100)

    # Hide the last empty subplot
    axes[2, 1].axis('off')

    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

plot_green_charts()
```
